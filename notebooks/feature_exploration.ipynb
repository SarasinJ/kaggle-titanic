```python
# get list of rooms per passenger, split by ' '
# 'Cabin' column has a list of rooms per passenger, separated by spaces, sometimes with multiple deck letters
# convert to an array of deck, number tuples

# initial look does not show passengers with multiple decks and multiple room numbers


def parse_cabin(string):
    if pd.isna(string):
        return []
    tokens = string.split()
    result = []
    deck_map = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'T': 8}
    for token in tokens:
        try:
            deck = deck_map[token[0]]
            room = int(token[1:])
        except (KeyError, ValueError): # exit the loop if we hit an unexpected format, 'F G67' for example
            return []
        result.append({'deck': deck, 'room': room})
    return result


train_df['rooms'] = train_df['Cabin'].apply(parse_cabin)
train_df['room_count'] = train_df['rooms'].apply(lambda x: len(x) if len(x) > 0 else np.nan)
train_df['deck'] = train_df['rooms'].apply(lambda x: x[0]['deck'] if len(x) > 0 else np.nan)


train_df
```
